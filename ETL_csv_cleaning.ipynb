```python
# Import dependencies 
import pandas as pd
```

```python
# Read in CSV file
users_df = pd.read_csv("Resources/BX-Users.csv", on_bad_lines='skip', encoding='latin-1', delimiter=';')
# Check to see if DataFrame read in correctly
users_df
```

```python
# Remove any null values from the dataframe
cleaned_users_df = users_df.dropna()
cleaned_users_df
```

```python
# Rename columns
renamed_users_df = cleaned_users_df.rename(columns={'User-ID':'User_id'})
renamed_users_df
```

```python
# Check data types of the DataFrame
renamed_users_df.dtypes
```

```python
# Change data types to desired type
refined_users_df = renamed_users_df.copy()
refined_users_df['Age'] = refined_users_df['Age'].astype('int64')
print(refined_users_df.dtypes)
refined_users_df
```

```python
# Export Data Frame into a new CSV
refined_users_df.to_csv("Resources/Users.csv", index=False, encoding='latin-1')
```

```python
# Read in second CSV for cleaning
books_df_new = pd.read_csv("Resources/BX-Books.csv", on_bad_lines='skip', encoding='latin-1', delimiter=';')
# Check that CSV was read in correctly
books_df_new
```

```python
# Remove any Null values from our DataFrame
cleaned_books_df_new = books_df_new.dropna()
cleaned_books_df_new
```

```python
renamed_books_df = cleaned_books_df_new.rename(columns={'Book-Title':'Book_Title', 'Book-Author':'Book_Author', 'Year-Of-Publication':'Publishing_Year'})
renamed_books_df
```

```python
# Drop any unneeded columns from our DataFrame
columns_to_drop = ['Image-URL-S', 'Image-URL-M', 'Image-URL-L']
# Create new dataframe with dropped values
cleaned_books_df = renamed_books_df.drop(columns=columns_to_drop)
cleaned_books_df
```

```python
# Check data types of our DataFrame
cleaned_books_df.dtypes
```

```python
# Assign new data types to fit our task
refined_books_df = cleaned_books_df.copy()
new_dtypes = {
    'ISBN': str,
    'Book_Title': str,
    'Book_Author': str,
    'Publishing_Year': int,
    'Publisher': str
}

refined_books_df = refined_books_df.astype(new_dtypes)
print(refined_books_df.dtypes)
refined_books_df
```

```python
# Export Data Frame into a new CSV
refined_books_df.to_csv("Resources/Books.csv", index=False, encoding='latin-1')
```
